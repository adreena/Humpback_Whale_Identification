```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import os
```

# Reading and viewing data

```python
train_df = pd.read_csv("../input/train.csv")
img = plt.imread('../input/train/{}'.format(train_df.iloc[0,0]))
plt.imshow(img)
train_df.head(5)
```

```python
# how many categories of whales ?
all_whales = train_df.Id.unique()
whales_distributions = train_df.groupby('Id')['Image'].nunique()
whales_distributions = whales_distributions.sort_values(ascending=True)

print('Number of all whales:', len(all_whales))
whales_distributions.plot(kind='line', figsize=(20,10), legend=True)
plt.show()

print('without the max:')
whales_distributions[:-1].plot(kind='line', figsize=(20,10), legend=True)

whales_distributions[.last(offset=10)]
```

```python
stats = whales_distributions.describe()
print('Categories of whales:', stats['count'])
print('Max {} number of images of whales belong to {}'.format(stats['max'], whales_distributions[-1]))
stats
```
